```julia
using Gadfly
using DataFrames
```

```julia
mat_watts = readdlm("out_watts")
categ = ["S","R","None"]
df = DataFrame(Number_of_winners = [mat_watts[j,i] for i=3:5 for j=1:size(mat_watts,1)], strain = [i for i in categ for j=1:size(mat_watts,1)])
```

```julia
p = plot(df, x = "Number_of_winners", color = "strain", Geom.histogram, Guide.title(" WS 10k Simulation winners"))
display(p)
```

```julia
Ndata_watts = DataFrame(S_Density = mat_watts[:,1],R_Density = mat_watts[:,2] ,Nwin_Proportion = [mat_watts[i,3]/mat_watts[i,4] for i = 1:size(mat_watts,1)], ClustC = mat_watts[:,7])
```

```julia
p = plot(Ndata_watts, x = "ClustC", y = "Nwin_Proportion", Guide.title("Susceptibles/Resistant by Network Global Clustering Coefficient"), Guide.ylabel("Proportion S/R"), Guide.xlabel("Global Clustering Coefficient"))
```

```julia
draw(SVG("Plots/CC_vs_Nnumber.svg", 10inch, 10inch), p)
```
