```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
mydata = np.genfromtxt("out_watts",delimiter="\t")
```

```python
clustc = [i[3] for i in mydata]
S_R_ratio = [i[0]/i[1] for i in mydata]
Res_win = [i[1] for i in mydata]
```

```python
%matplotlib inline
plt.plot(S_R_ratio,clustc,'ro')
plt.xlabel("S_R Ratio")
plt.ylabel("Clustering Coefficient")
plt.annotate('N=50,A=40,clen=0.2', xy=(0.95, 0.85), xycoords='axes fraction')
plt.show()
```

```python

def count(a,b):
    miny  = round(min(b),3)
    minx = round(min(a),3)
    
    maxy = round(max(b),3)
    maxx = round(max(a),3)
    
    x = np.linspace(minx,maxx,(maxx-minx)*100+1)
    y = np.linspace(miny,maxy,(maxy-miny)*100+1)

    counts = np.zeros((len(y),len(x)))
    min_x = int(round(100*min(a)))
    min_y = int(round(100*min(b)))
    a1= [int(round(100 * i)) for i in a]
    b1= [int(round(100 * i)) for i in b]
    
    for i in range(len(a)):
        counts[b1[i]-min_y,a1[i]-min_x] +=1
        
    freq_vec = [0] * len(a)
    for i in range(len(a)):
        freq_vec[i]=counts[b1[i]-min_y,a1[i]-min_x]
    return [x,y,counts,freq_vec] 
```

```python
def obtain_freqvec(a,b):
    miny  = min(b)
    minx = round(min(a),3)
    
    maxy = max(b)
    maxx = round(max(a),3)
    
    x = np.linspace(minx,maxx,(maxx-minx)*100+2)
    y = np.arange(miny,maxy+1,1)

    counts = np.zeros((len(y),len(x)))
    min_x = int(round(100*min(a)))
    min_y = min(b)
    a1= [int(round(100 * i)) for i in a]
    b1= [i for i in b]

    for i in range(len(a)):
        counts[b1[i]-min_y,a1[i]-min_x] +=1

    freq_vec = [0] * len(a)
    for i in range(len(a)):
        freq_vec[i]=counts[b1[i]-min_y,a1[i]-min_x]
    return freq_vec 
```

```python
R_freq_vec = obtain_freqvec(clustc,Res_win)
```

```python
x,y,z,freq_vec = count(clustc,S_R_ratio)
x, y = np.meshgrid(x, y)
```

```python
print(len(x),len(y),len(z))
```

```python
%matplotlib inline
plt.scatter(clustc,Res_win, c = R_freq_vec)
plt.gray() 
cbar = plt.colorbar(cmap=cm.gray(max(R_freq_vec)-min(R_freq_vec)))
cbar.set_label('Frequency')
plt.title("Resistance Selection on Clustering Coefficient")
plt.ylabel("Nodes won by R")
plt.xlabel("Global Clustering Coefficient")
plt.annotate('N=50,A=40,clen=0.2', xy=(0.65, 0.9), xycoords='axes fraction')
plt.savefig("plots/S_CC_Freqs.png")
plt.show()
```

```python
plt.scatter(clustc,S_R_ratio, c = freq_vec)
plt.gray() 
cbar = plt.colorbar(cmap=cm.gray(max(freq_vec)-min(freq_vec)))
cbar.set_label('Frequency')
plt.title("Resistance Selection on Clustering Coefficient")
plt.ylabel("S/R Winner Ratio ")
plt.xlabel("Global Clustering Coefficient")
plt.annotate('N=50,A=40,clen=0.2', xy=(0.95, 0.85), xycoords='axes fraction')
plt.show()
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

surf = ax.plot_surface(x,y,z,cmap=cm.coolwarm,
                       linewidth=0.1)

fig.colorbar(surf, shrink=0.5, aspect=5)

ax.set_title('Node winners by clustering coefficient')

xLabel = ax.set_xlabel('\nClustering Coefficient')
yLabel = ax.set_ylabel('S/R Winners Ratio')
zLabel = ax.set_zlabel('\nFrequency')
ax.view_init(azim=30)

plt.show()
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

surf = ax.plot_surface(x[2:],y[2:],z[2:],cmap=cm.coolwarm,
                       linewidth=0)

fig.colorbar(surf, shrink=0.5, aspect=5)

ax.set_title('Node winners by clustering coefficient (without first row)')

xLabel = ax.set_xlabel('\nClustering Coefficient')
yLabel = ax.set_ylabel('S/R Ratio')
zLabel = ax.set_zlabel('\nFrequency')
ax.view_init(azim=120)

plt.show()
```
